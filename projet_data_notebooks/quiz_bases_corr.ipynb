## 1. Opérateurs de calcul

```python
# Q1 — Division entière et modulo
# Combien de semaines complètes et de jours restants dans 100 jours ?
jours = 100
semaines = jours // 7          # division entière
reste    = jours % 7          # jours restants

assert semaines == 14
assert reste    == 2
```

```python
# Q2 — Puissance
# 2 à la puissance 10
resultat = 2 ** 10

assert resultat == 1024
```

```python
# Q3 — Division réelle vs entière
# 7 divisé par 2 doit donner un float
r_float = 7 / 2
r_int   = 7 // 2           # partie entière uniquement

assert type(r_float) is float and r_float == 3.5
assert type(r_int)   is int   and r_int   == 3
```

```python
# Q4 — Concaténation et répétition de chaînes
a = "bon"
b = "jour"
mot    = a + b            # concatène a et b
triple = " ".join([mot] * 3)            # répète mot 3 fois

assert mot    == "bonjour"
assert triple == "bonjour bonjour bonjour"
```

```python
# Q5 — Priorité des opérateurs
# Calcule 3 + 4 * 2
r1 = 3 + 4 * 2          # priorité à la multiplication
r2 = (3 + 4) * 2          # priorité à l'addition

assert r1 == 11
assert r2 == 14
```

## 2. Opérateurs de comparaison et logique

```python
# Q6 — Comparaisons chaînées
age = 25
# Est-ce que age est strictement entre 18 et 65 ?
majeur_actif = 18 < age < 65

assert majeur_actif is True
```

```python
# Q7 — Opérateurs logiques
x, y = True, False
r1 = x and y                # x et y
r2 = x or y                # x ou y
r3 = not x               # non x

assert r1 is False
assert r2 is True
assert r3 is False
```

```python
# Q8 — Valeur booléenne des types
# Complète avec True ou False
assert bool(0)    == False
assert bool("")   == False
assert bool([])   == False
assert bool(None) == False
assert bool("0")  == True
```

```python
# Q9 — is vs ==
a = [1, 2, 3]
b = [1, 2, 3]
c = a

assert (a == b) == True
assert (a is b) == False
assert (a is c) == True # referencent le même objet
```

```python
# Q10 — Comparaison de chaînes (ordre lexicographique Unicode)
assert ("abc" < "abd")  == True
assert ("ABC" < "abc")  == True
assert ("zoo" > "Zoo")  == True
```

## 3. Opérateurs in-place

```python
# Q11 — Compteur
compteur = 0
compteur += 1         # incrémente de 1
compteur += 1         # encore une fois
compteur *= 3         # multiplie par 3

assert compteur == 6
```

```python
# Q12 — In-place sur liste
l = [1, 2]
ref = l           
l += [3, 4]           # étend la liste existante

assert l   == [1, 2, 3, 4]
assert ref is l         # même objet
```

## 4. Opérateur `[]` — index, slice, clé

```python
# Q14 — Index positif et négatif
villes = ['Toulouse', 'Bordeaux', 'Lyon', 'Paris', 'Lille']

premiere = villes[0]          # premier élément
derniere  = villes[-1]         # dernier élément sans connaître la longueur
avant_der = villes[-2]         # avant-dernier

assert premiere == 'Toulouse'
assert derniere  == 'Lille'
assert avant_der == 'Paris'
```

```python
# Q15 — Slices
villes = ['Toulouse', 'Bordeaux', 'Lyon', 'Paris', 'Lille']

deux_premiers = villes[:2]     # les 2 premiers
deux_derniers = villes[-2:]     # les 2 derniers
pairs         = villes[::2]     # éléments aux indices pairs (0, 2, 4)
inverse       = villes[::-1]     # liste inversée

assert deux_premiers == ['Toulouse', 'Bordeaux']
assert deux_derniers == ['Paris', 'Lille']
assert pairs         == ['Toulouse', 'Lyon', 'Lille']
assert inverse       == ['Lille', 'Paris', 'Lyon', 'Bordeaux', 'Toulouse']
```

```python
# Q16 — Slice de chaîne
city = "L'Haÿ-les-Roses"

cinq_premiers = city[:5]     # 5 premiers caractères
cinq_derniers = city[-5:]     # 5 derniers caractères

assert cinq_premiers == "L'Haÿ"
assert cinq_derniers == "Roses"
```

```python
# Q17 — Accès dans un dict
ville = {'nom': 'Toulouse', 'population': 504_078, 'dep': '31'}

nom = ville['nom']               # nom de la ville
pop = ville['population']               # population

assert nom == 'Toulouse'
assert pop == 504_078

# Accès avec valeur par défaut si clé absente
region = ville.get('region', 'Inconnue')
assert region == 'Inconnue'
```

## 5. Builtins : `len`, `str`, `repr` et conversions

```python
# Q18 — len
assert len('hello')        == 5
assert len([1, 2, 3])      == 3
assert len((1, 2))         == 2
assert len({'a': 1, 'b': 2}) == 2
assert len([[1, 2, 3], [4, 5, 6]]) == 2
assert len([{'a': 1, 'b': 2}, {'a': 3, 'b': 4}]) == 2
assert len({'a': [1, 2, 3], 'b': [4, 5, 6]}) == 2
```

```python
# Q20 — Conversions
assert int('42')      == 42
assert int('ff', 16)  == 255   # hexadécimal
assert float('3.14')  == 3.14
assert bool(0)        == False
assert bool('false')  == True
assert str(True)      == 'True'
assert list('abc')    == ['a', 'b', 'c']
assert list((1, 2, 3)) == [1, 2, 3]
```

## 6. Tri : `sort`, `sorted`, `locale.strxfrm`

```python
# Q21 — sorted vs sort
nombres = [3, 1, 4, 1, 5, 9, 2, 6]

# trier avec sorted()
tries = sorted(nombres)             # liste triée
assert tries   == [1, 1, 2, 3, 4, 5, 6, 9]
assert nombres == [3, 1, 4, 1, 5, 9, 2, 6]  # inchangé !

# trier avec sort()
retour = nombres.sort()
assert retour  is None
assert nombres == [1, 1, 2, 3, 4, 5, 6, 9]  # modifié
```

```python
# Q22 — Tri décroissant et par clé
villes = ['Toulouse', 'bordeaux', 'Lyon', 'paris']

# Tri décroissant
decr = sorted(villes, reverse=True)
assert decr == ['paris', 'bordeaux', 'Toulouse', 'Lyon']  # ordre ASCII décroissant

# Tri insensible à la casse
sans_casse = sorted(villes, key=str.lower)
assert sans_casse == ['bordeaux', 'Lyon', 'paris', 'Toulouse']
```

```python
# Q23 — Tri selon les règles de la locale (accents, casse)
locale.setlocale(locale.LC_ALL, 'fr_FR.UTF-8')  # locale français de France

villes = ['Évreux', 'angers', 'épinal', 'Bordeaux', 'brest', 'Zutkerque', 'Œuilly']

# Tri selon les règles linguistiques de la locale courante
tries_locale = sorted(villes, key=locale.strxfrm)

assert tries_locale == ['angers', 'Bordeaux', 'brest', 'épinal', 'Évreux', 'Œuilly', 'Zutkerque']
```

## 7. f-strings

```python
# Q24 — f-string de base
nom = 'Alice'
age = 30
msg = f"{nom} a {age} ans"               # f-string produisant : "Alice a 30 ans"

assert msg == 'Alice a 30 ans'
```

```python
# Q25 — Formatage numérique dans une f-string
population = 504_078
prix = 3.14159

# Nombre entier avec séparateur de milliers et 2 décimales pour le flottant
s1 = f"{population:_}".replace('_', ' ')                # ex: '504 078'  (séparateur espace)
s2 = f"{prix:0.2f}"                # ex: '3.14'     (2 décimales)
assert s1 == '504\u202f078' or s1 == '504 078'   # espace ou espace fine
assert s2 == '3.14'
```

```python
# Q26 — Alignement et remplissage
# Affiche un index sur 2 chiffres avec zéro à gauche, puis la ville
i, ville = 3, 'Lyon'
ligne = f"{i:02d} - {ville}"             # produit : '03 - Lyon'

assert ligne == '03 - Lyon'
```

## 8. Types temporels `datetime` + fuseaux horaires IANA

```python
# Q27 — Créer une date et une datetime
# Date du 14 juillet 2025
fete_nat = date(2025, 7, 14)

assert type(fete_nat) is date
assert fete_nat.year  == 2025
assert fete_nat.month == 7
assert fete_nat.day   == 14

# Datetime naive : 14 juillet 2025 à 11h00 pile
defile = datetime(2025, 7, 14, 11)
assert type(defile) is datetime
assert defile.date() == fete_nat
assert defile.hour == 11 and defile.minute == 0
```

```python
# Q28 — timedelta : durée entre deux dates
debut  = date(2025, 1,  1)
fin    = date(2025, 12, 31)
duree  = fin - debut

assert duree.days == 364
```

```python
# Q29 — timedelta : ajout d'une durée
debut  = datetime(2026, 6,  30, 9)
duree = timedelta(days=2, hours=8)             # ajouter 2 jours et 8 heures
fin   = debut + duree
print(fin)
assert fin == datetime(2026, 7,  2, 17)
```

```python
# Q30 — Parser une chaîne ISO 8601 en datetime
s = '2025-07-14T11:00:00'
d = datetime.fromisoformat(s)   # Hint: utiliser méthode `fromisoformat`

assert d.year == 2025 and d.hour == 11
assert d.tzinfo is None   # naive : pas de fuseau horaire
```

```python
# Q31 — Fuseau horaire IANA
paris_tz  = ZoneInfo('Europe/Paris')         # Europe/Paris
tokyo_tz  = ZoneInfo('Asia/Tokyo')         # Asia/Tokyo

# Datetime aware : 14 juillet 2025 11h00 heure de Paris
defile_paris = datetime(2025, 7, 14, 11, tzinfo=paris_tz)

# Convertir en heure de Tokyo
defile_tokyo = defile_paris.astimezone(tokyo_tz)      # Hint: utiliser méthode `astimezone`

assert type(paris_tz) is ZoneInfo and str(paris_tz) == 'Europe/Paris'
assert type(tokyo_tz) is ZoneInfo and str(tokyo_tz) == 'Asia/Tokyo'
assert defile_tokyo.hour == 18   # Tokyo est UTC+9, Paris UTC+2 en été → +7h
assert str(defile_tokyo.tzinfo) == 'Asia/Tokyo'
```

```python
# Q32 — Formater une datetime en chaîne
d = datetime(2025, 7, 14, 11, 0, 0)

# Format cible : '14/07/2025 11:00'
s = d.strftime('%d/%m/%Y %H:%M')

assert s == '14/07/2025 11:00'
```

## Bonus — Pandas : `.iloc`, `.loc` et prédicats
On utilise les villes françaises de plus de 100 000 habitants.

```python
df = pd.read_json('data/villes100k.json', dtype={'dep_code': str, 'code_postal': str})
df.head(3)
```

```python
# Q33 — .iloc : accès par position entière
# Première ligne (indice 0), toutes les colonnes
ligne0 = df.iloc[0]            
assert ligne0['nom_standard'] == 'Nice'

# 3 premières lignes, colonnes 0 et 2 (nom_standard et dep_code)
extrait = df.iloc[:3, [0, 2]]          
assert list(extrait.columns) == ['nom_standard', 'dep_code']
assert len(extrait) == 3
```

```python
# Q34 — .loc : accès par label / prédicat
# Toutes les villes du département 69
dep69 = df.loc[df.dep_code == '69']
assert set(dep69['nom_standard']) == {'Lyon', 'Villeurbanne'}

# Colonnes nom_standard et population uniquement
dep69_slim = dep69[['nom_standard', 'population']]
assert list(dep69_slim.columns) == ['nom_standard', 'population']
assert len(dep69_slim) == 2
```

```python
# Q35 — Prédicat combiné
# Villes de + de 300 000 hab. OU dans le département 13
grands_ou_13 = df.loc[(df.population >= 300_000) | (df.dep_code == '13')]      
assert 'Marseille'   in grands_ou_13['nom_standard'].values
assert 'Paris'       in grands_ou_13['nom_standard'].values
assert 'Aix-en-Provence' in grands_ou_13['nom_standard'].values

# Population minimale de ces villes
pop_min = grands_ou_13.population.min()
assert pop_min == 147_478   # Aix-en-Provence
```

```python
# Q36 — .loc avec prédicat str
# Villes dont le nom commence par 'S'
commence_s = df.loc[df.nom_standard.str.startswith('S')]
assert all(n.startswith('S') for n in commence_s['nom_standard']) # pour vérifier sans Pandas 😅
print(commence_s[['nom_standard', 'dep_code']])
```
