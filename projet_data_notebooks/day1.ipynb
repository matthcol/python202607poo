```python
def incr_city(list_city_pop: list[tuple[str, int]], index: int, delta: int) -> None:
    name, pop = list_city_pop[index]
    list_city_pop[index] = (name, pop + delta)
```

```python
def is_palindrome(word: str) -> bool:
    return word.lower() == word[::-1].lower()
```

### Expression `for` vs `map/reduce`

```python
[city for city in cities if city.strip().lower().startswith('b')]
```

```python
[city.strip().lower() for city in cities]
```

```python
cities_filtered = [
    city_n for city_n in (
        city.strip().lower() for city in cities
    ) if city_n.startswith('b')
]
cities_filtered
```

```python
numbers = [12, 23, 67, 43]
[n**2 + 1 for n in numbers]
```

```python
[n**2 + 1 for n in numbers if n % 2 == 1]
```

```python
g = (n**2 + 1 for n in numbers if n % 2 == 1)
g
```

```python
sum(g)
```

```python
list(
    filter(
        lambda city_n: city_n.startswith('b'),
        map(
            lambda city: city.strip().lower(),
            cities
        )
    )
)
```

```python
cities_500k = []
for city in data:
    if city['population'] >= 500_000:
        # print(city)
        cities_500k.append(city)

# pretty print
for city in cities_500k:
    print(city)
```

```python
cities_500k = [city for city in data if city['population'] >= 500_000]

# pretty print
for city in cities_500k:
    print(city)
```
